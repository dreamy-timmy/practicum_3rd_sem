```python
wanted_cabins = titanic['cabin'].dropna().unique()
wanted_cabins
```

```python
w = united['cabin'].dropna().unique()
w
```

Удостоверимся, что все каюты в первом датафрейме присутствуют и во втором:

```python
all(wanted_cabins[i] in w for i in range(len(wanted_cabins)))
```

```python
ans_united = []
ans_titanic = []
```

```python
for cabin in wanted_cabins:
    ans_united.append( united.query(f'cabin == "{cabin}"')['name'].value_counts() )
    ans_titanic.append( titanic.query(f'cabin == "{cabin}"')['name'].value_counts() )
    # print( united.query(f'cabin == "{cabin}"')['name'].value_counts() )
    # print('------------------------------------------------')
```

```python
for i in range(len(ans_titanic)):
    unit = len(ans_united[i].index)
    titan = len(ans_titanic[i].index)
    if unit > titan: 
        print(f'Люди из датафрейма "титаник": {[x for x in ans_titanic[i].index]}')
        print(f'Люди, плывшие в одной каюте с людьми из датафрейма "титаник": {[x for x in ans_united[i].index[unit-titan:]]}')
        # print(unit, titan)
        print('-----------------------------------------------------------')
```

Известно, что трагедия титаника случилась в 1912 г.:

```python
tragedy = 1912
BirthYear = []
for age in united['age']:
    # print(str(int(age)).isdigit()) 
    if str(age)[0].isdigit() and str(int(age)).isdigit(): BirthYear.append(tragedy - age)
    else: BirthYear.append(None)
```

```python
print(BirthYear)
```

```python
united['birthyear'] = BirthYear
```
