```python
fig, axs = plt.subplots(figsize=(10, 10))

unique_counts = np.unique(data['quality'], return_counts=True)

plt.pie(unique_counts[1], labels=unique_counts[0], autopct='%1.2f%%', colors=['#b2de21','#4c0d57','#574b0d','#0d5750'])
fig.savefig('3.png')
```

Тепловая карта

```python
plt.figure(figsize = (12,15))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm",fmt=".2f")
plt.savefig('4.png')
```

```python
f = plt.figure(figsize=(20,10))

def mapper(value):
  if value in [3, 4]:
    return "Low Quality" 
  elif value in [5, 6]: 
    return "Ok Quality"
  else: return "High Quality"

quality_mark = data["quality"].map(mapper)
sns.countplot(x = data["Density"].head(100), hue=quality_mark)
plt.savefig('5.png')
```

Проверка на нормальное распределение:

```python
import scipy as sp

a = np.array(data['Density'][:200])
print(sp.stats.normaltest(a)) 

# If the p-val is very small - unlikely that the data came from a normal distribution. 
```
